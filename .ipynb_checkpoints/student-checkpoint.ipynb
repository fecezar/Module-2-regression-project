We can see that all zip codes high price outliers in a relatively equal measure. Therefore outliers should be 'flagged' in relation to the price of the relevant zip code. 

```python
#

def groupby_outliers_subset(df, quantile_threshold = 0.97, 
                            no_outliers=True, 
                            groupping='zipcode',
                            metric='price'):

    outliers_indices_list = []
    for groupping, groupping_df in list(df.groupby(groupping)):
        metric_threshold = groupping_df[metric].quantile(quantile_threshold)
        outliers_indices =  list(
            groupping_df[groupping_df[metric] > metric_threshold].index)
        outliers_indices_list += outliers_indices
        
    if no_outliers:
        return df.drop(index=outliers_indices_list).reset_index(drop=True)
    else:
        return df.loc[outliers_indices_list].reset_index(drop=True)
```

### Multicollinearity 

In order to check for multicollinearity between independent variables, let's write a function that returns all variables with a correlation above a certain threshold. This will inform variable selection decisions later on

```python
# correlated variables


def create_colinearity_df(correlation_hurdle, variables_list, df):
    """Function that finds which variables have correlation above a 
    certain threshold.
    
    ***
    Function takes in as arguments a correlation hurdle, a list of variables 
    and a Data Frame.
    
    Returns a correlation matrix as a pandas Data Frame
    ***
    Arguments:
    - correlation_hurdle : float between 0 and 1
    - variables_list: list of strings with variable names. Must be a column 
    name of the provided data frame
    - df : a DataFrame containing the data for the variables.
    
    """
    
    # creates a new df from a correlation matrix
    # multi index structure 
    correlations_df = pd.DataFrame(
        df[variables_list].corr().stack()).reset_index()
    
    # removes redundant correlations
    index_drop = correlations_df[
        correlations_df['level_0'] == correlations_df['level_1']].index
    correlations_df.drop(index=index_drop, inplace=True)

    # renames columns, resets index
    correlations_df.columns= ['x', 'y', 'corr']
    correlations_df.reset_index(inplace=True)
    correlations_df.drop(columns='index', inplace=True)
    
    # creates new column with the correlated pairs
    correlations_df['pair'] = list(zip(correlations_df['x'], correlations_df['y']))
    
    # drops columns used to create the pairs' column
    correlations_df.drop(columns=['x', 'y'], inplace=True)
    
    # sets the pairs as new index
    correlations_df.set_index('pair', inplace=True)
    
    # drops duplicates
    correlations_df.drop_duplicates(inplace=True)

    
    # returns only pairs with correlation above hurdle
    return correlations_df[abs(correlations_df['corr']) > correlation_hurdle]

create_colinearity_df(0.75, no_outliers_df.drop(columns='price').columns, no_outliers_df)
```

To aid in deciding which variables to drop, let's create a function to find a simple r-squared
for each variable, individually. 

```python
def create_rsquared_df(dep_variable, ind_variables, df):
    rsqs = []
    variables = []
    
    for variable in ind_variables:
        f = str(dep_variable)+'~'+str(variable)
        model = ols(formula=f, data=df).fit()
        rsqs.append(model.rsquared)
        variables.append(variable)
        
    return pd.DataFrame(dict(r_squared_ = rsqs), index=variables)

create_rsquared_df('price', no_outliers_df.drop(columns='price').columns, no_outliers_df)
```

- **avg_neighbor_price**: common sense tells us that prices in the vicinity of a property are correlated to the price of the property itself. Therefore we should calculate the average price of the close neighbors of a property  

The first step is to find the nearest neighbors. We can do that with the coordinates from latitude and longitude

```python

# creates a coordinates column 

df['coord'] = list(zip(df['lat'], df['long']))

# updates no_outliers_df with  coord
no_outliers_df = groupby_outliers_subset(df, 
                                         quantile_threshold = 0.97, 
                                         no_outliers=True, 
                                         groupping='zipcode',
                                         metric='price')

outliers_df = groupby_outliers_subset(df, 
                                      quantile_threshold = 0.97, 
                                      no_outliers=False, 
                                      groupping='zipcode', 
                                      metric='price')


# function that finds avg price of a list of houses sold based on 
# an array with their of indices

def nearest_neighbors_prices(n, coord_array, df):
    """ Function takes in a 'n' number of neighbors
    and an array of coordinates.
    
    Returns a one dimentional Data Frame with average price 
    of n nearest neighbors
    
    n : number of neighbors to compute in the average price
    coord_array : array with the coordinates for which the function finds the 
    nearst neighbors
    """
    
    # creates a NearestNeighbor object
    nbrs  = NearestNeighbors(n_neighbors=n, #number of neighbors to compute
                             metric='euclidean', 
                             algorithm='kd_tree').fit(coord_array)
    
    # the results include the distances of the neighbors and 
    # a list of n indices (i.e. location in coord_array) 
    distances, indices = nbrs.kneighbors(coord_array)
    
    mean_neighbors_price_list = []
    
    # parses the returned indices
    for lst in indices:
        
        # the first is the index of the coordinats for which 
        # we want to find neighbors
        index = lst[0]  
        
        
        # the others are the neighbors'
        neighbors_indices = lst[1:]
        
        # calculates the average price by fetching values from 
        # df['price']
        mean_neighbors_price = float(df.loc[neighbors_indices, 
                                       ['price']].mean())
        

        mean_neighbors_price_list.append(mean_neighbors_price)

    return pd.DataFrame(dict(
        mean_neighbors_price = mean_neighbors_price_list))
```

Next step is to chose a value for **n** that maximizes the r-squared for our new feature

```python

for dataframe in [outliers_df, no_outliers_df]:
    best_rsquared = 0
    best_n = 0
    print('*********  ******')
    
    # loop to test the value for n that maximizes 
    # r squared for avg_neighbor_price 
    for n in range(4, 10, 1): 
        coord_array = np.array(list(dataframe.coord))
        
        # creates a avg_neighbor_price column with a given n
        dataframe['avg_neighbor_price'] = nearest_neighbors_prices(n, 
                                                                   coord_array,
                                                                  dataframe)

        # finds r squared given n  
        f = 'price~avg_neighbor_price' 
        model = ols(formula=f, data=dataframe).fit()
        rsquared = model.rsquared

        print('*****')
        print('n = ', n)
        print('rsquared = ', rsquared)
        
        # stores n if it is the best so far 
        if rsquared > best_rsquared:
            best_rsquared = rsquared
            best_n = n
        else:
            continue

    # with the best n found, recreates the avg_neighbor_price column
    dataframe['avg_neighbor_price'] = nearest_neighbors_prices(best_n, 
                                                               coord_array, 
                                                               dataframe)


    print('*****')
    print('*****')
    print('best n = ', best_n)
    print('best rsquared = ', best_rsquared)

    # best n =  7
    # best rsquared =  0.5939435877909922
```

# Variable Selection and Model Training

 Let's start by selecting variables in order to Adjusted maximize R Squared, so we have a model that is as accurate as possible.

## Forward Selection for Adj R-Squared 
The function below will find all the variables that contribute to Adjusted R Squared.



```python
# forward selection of variables to maximixe r-squared

def forward_selected(data, response):
    """Linear model designed by forward selection.

    Parameters:
    -----------
    data : pandas DataFrame with all possible predictors and response

    response: string, name of response column in data

    Returns:
    --------
    model: an "optimal" fitted statsmodels linear model
           with an intercept
           selected by forward selection
           evaluated by adjusted R-squared
    """
    remaining = set(data.columns)
    remaining.remove(response)
    
    #
    print(len(remaining))
    
    selected = []
    current_score, best_new_score = 0.0, 0.0
    while remaining and current_score == best_new_score:
        scores_with_candidates = []
    
        
        for candidate in remaining:
            formula = "{} ~ {} + 1".format(response,
                                           ' + '.join(selected + [candidate]))
            score = smf.ols(formula, data).fit().rsquared_adj
            scores_with_candidates.append((score, candidate))
        scores_with_candidates.sort()
        best_new_score, best_candidate = scores_with_candidates.pop()
        if current_score < best_new_score:
            remaining.remove(best_candidate)
            selected.append(best_candidate)
            current_score = best_new_score
    formula = "{} ~ {} + 1".format(response,
                                   ' + '.join(selected))
    model = smf.ols(formula, data).fit()
    return model
```

## Prediction Model Training
The next function will test different splits for test-train and see which ones have the best results as measured by the MSE. 
Additionally, the function will purge all variables with a p value above 5%

```python


def model_training(data,
                   shares_to_test, 
                   alpha= 0.05, 
                   random_state=4, 
                   filename='best_model'):
    
    # creates data to be separated into train and test
    X = data.drop(columns='price')
    y = data['price']
    
    
    alpha_test = 1
    while alpha_test > 0:




        cv_results_list = []
        best_model = None

        # loop overs different train-test divisions to find the one 
        # with lowest mean squared error

        for share in shares_to_test:

            # creates train and test data
            X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                                test_size=share, 
                                                                random_state=4)
            # training data
            training_data = pd.concat([X_train, y_train], axis=1)
            response = 'price'

            # Forward selection using adjusted R-squared (on train data)
            candidate_model = forward_selected(training_data, response)

            # stores the selected variables
            candidate_variables = list(pd.DataFrame(candidate_model.params).index)
            candidate_variables.remove('Intercept')


            # cross validation 
            linreg = LinearRegression()    
            cv_results = cross_val_score(linreg, 
                                         X[candidate_variables], 
                                         y, 
                                         cv=5, 
                                         scoring='neg_mean_squared_error')

            # stores the mean squared error
            model_mean_squared_error = cv_results.mean()
            cv_results_list.append(model_mean_squared_error)

            # stores the model if the cross validation result is the best so far
            if max(cv_results_list) <= model_mean_squared_error:
                best_model = candidate_model
                best_share = share*100
            else:
                continue
            
            
    # 
        #
            #
        # saves best model


    
        pvalues_df = pd.DataFrame(best_model.pvalues)
        pvalues_df.columns=['pvalue'] 

        alpha_test = (pvalues_df['pvalue'] > alpha).sum()
        print('alpha_test : ', alpha_test)
        
        if alpha_test > 0:
            filt = pvalues_df['pvalue'] > alpha
            columns_to_drop = list(pvalues_df.loc[filt].index)
            X.drop(columns=columns_to_drop, inplace=True)
        else:
            continue
        
        
    file_name = 'best_prediction_model.pickle' 
    best_model.save(file_name)
    print(file_name, ' saved')
    print('optimal share: ', best_share)
        
    return best_model

```

```python
# variables to consider (after removing correlated independent variables)
model_df = pd.concat((list(model_variables.values()) + [no_outliers_df.price]), axis=1)

# final model as calculated by the functions above 

best_prediction_model = model_training(model_df,[0.30, 0.25, 0.20])#, 0.25, 0.20])
best_prediction_model.summary()
```

Now let's select variables to make **inferences**. In this variable selection strategy we will only select the 'n' best variables

The first step is to create a function that returns the top n variables

```python

def top_variables(n_features_to_select, 
                  test_size, 
                  data,
                  random_state=4):
    """Function to find the n best predictors
    
    Arguments:
    - n_features_to_select : int, number of features to be returned
    - test_size : float, share of the data to be used for testing
    - data : dataframe containing the entire data set, including testing data, 
    predictors and dependend variable
    - random_state : optional. if not provided, default is 4"""
    
    # creates data to be separated into train and test
    X = data.drop(columns='price')
    y = data['price']

    # creates train and test data
    X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                        test_size=test_size, 
                                                        random_state=4)
    # fits a linear regression to the provided date using the n best features
    linreg = LinearRegression()
    selector = RFE(linreg, n_features_to_select=n_features_to_select)
    selector = selector.fit(X_train, y_train)
    
    # loops an array of column names and an array of boolean values that indicate
    # which variables have been selected
    top_variables = []
    for i, (col, test) in enumerate(zip(X_train.columns, # varibles considered
                                        selector.support_ # array of bool
                                       )):
        if test:
            
            top_variables.append(col)
    
    estimators = selector.estimator_
    top_variables_coef = estimators.coef_
    
    
    formula = 'price ~ %s'%("+".join(top_variables))
    lin_reg = smf.ols(formula, 
                      data=pd.concat([X_train, y_train], axis=1)
                     ).fit()
    
    err_series = lin_reg.params - lin_reg.conf_int()[0]
    
    coef_df = pd.DataFrame({'coef': lin_reg.params.values[1:],
                        'err': err_series.values[1:],
                        'varname': err_series.index.values[1:]
                       })
    
    
   # top_var_df = pd.DataFrame(dict(
   #     coefs = top_variables_coef, 
   #     variable = top_variables))
    
    return coef_df
```
